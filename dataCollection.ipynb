```python
import requests 
import pandas as pd
```

```python
company_url = "https://data.sec.gov/files/company_tickers.json"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36",
    "Accept": "application/json",
    "Accept-Encoding": "gzip , deflate",
    "host": "www.sec.gov",
}

data = requests.get(company_url, headers=headers)
# from json to data frame
df = pd.DataFrame.from_dict(data.json(), orient="index")
df.cik_str = df.cik_str.astype(str).str.zfill(10)
df["sicDescription"] = None

for index, row in df.head(10).iterrows():
    cik = row["cik_str"]
    url = f'https://data.sec.gov/submissions/CIK{cik}.json'

    try:
        sub_response = requests.get(url, headers=headers)
        submissions = sub_response.json()
        df.at[index, "sicDescription"] = submissions.get("sicDescription", None)


    except requests.exceptions.RequestException:
        df.at[index, "sicDescription"] = None

df.to_csv("./data/company_tickers.csv", index=False)
```
